##冪函數與算術平均數的應用:影像亮度設計與調整


##形如y=x^a(a為常數)的函數稱為冪函數Power function

```python
#冪函數(power function)
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5),dpi=80)
plt.title("power function")   #標題名稱

x=np.linspace(0,1,100)  #(起點,終點,共幾個點包含端點)
y1=x**3
y2=x**2
y3=x
y4=x**(1/2)
y5=x**(1/3)

plt.plot(x,y1,color="blue",linewidth=2)
plt.plot(x,y2,color="red",linewidth=2)
plt.plot(x,y3,color="green",linewidth=2)
plt.plot(x,y4,color="yellow",linewidth=2)
plt.plot(x,y5,color="brown",linewidth=2)

ax=plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.show()
```

##實作1:使用凹函數:

```python
import matplotlib.pyplot as plt
from PIL import Image
plt.figure(figsize=(8,8),dpi=80)
pic=Image.open("A.jpg")
m,n=pic.size
new=Image.new("RGB",(m,n),(255,255,255))
def f(x):
  return x**0.7
for i in range(m):
  for j in range(n):
    r,g,b=pic.getpixel((i,j))
    R=255*f(r/255)
    G=255*f(g/255)
    B=255*f(b/255)
    new.putpixel((i,j),(int(R),int(G),int(B)))
new.save("A1.jpg")

image=Image.open("A1.jpg")
plt.imshow(image)
```

##實作5:使用算術平均數

```python
import matplotlib.pyplot as plt
from PIL import Image
plt.figure(figsize=(8,8),dpi=80)
pic=Image.open("A.jpg")
m,n=pic.size
new=Image.new("RGB",(m,n),(255,255,255))

for i in range(m):
  for j in range(n):
    r,g,b=pic.getpixel((i,j))
    R=(r+g+b)/3
    G=(r+g+b)/3
    B=(r+g+b)/3
    new.putpixel((i,j),(int(R),int(G),int(B)))
new.save("A5.jpg")

image=Image.open("A5.jpg")
plt.imshow(image)
```
