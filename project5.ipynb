#專題:二元一次聯立不等式的圖解與格子點

##基本範例:
##圖示二元一次聯立不等式 x-y-3<=0 與 2x+y-2>0

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(6,6),dpi=80) #視情況可調整
(a,b)=(-5,10) #設定橫座標範圍(需調整)
(c,d)=(-7,8)  #設定縱座標範圍(需調整)

x=np.linspace(a,b,2) #(起點,終點,共幾個點包含端點)
y1=x-3   #直線L1(需先整理為y=mx+k型式)
y2=-2*x+2  #直線L2(需先整理為y=mx+k型式)
plt.plot(x,y1,color="blue",linestyle="-",linewidth=3) #實線
plt.plot(x,y2,color="blue",linestyle=":",linewidth=3) #虛線

m=2 #調整塗色的密度,若m=1則顯示格子點
for x in range(a*m,b*m+1,1):
  x=x/m
  for y in range(c*m,d*m+1,1):
    y=y/m
    if(x-y-3<=0 and 2*x+y-2>0): #可去掉等號,看其效果如何?
      plt.scatter(x,y,color="pink",marker="x")

ax = plt.gca()  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xlim(a,b)
plt.ylim(c,d) 
plt.show()
```
